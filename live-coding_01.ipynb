## Dopna & Fillna

- 'dropna()' entfernt alle Zeilen mit NaN
- mit dropna(axis=1) kann man alle Spalten entfernen die NaN enthalten
- 'fillna()' ersetzt alle NaN mit einem Wert

```python
data_selected.dropna().info() #zu krass, da alle unnoetigen spalten rausfliegen
```

```python
mask_type2 = data_selected['type_2'].isna()
mask_type2
```

```python
data_cleaned = data_selected.loc[~mask_type2] #Tilde negiert die Maske und gibt nur die Zeilen aus, die nicht in der Maske sind
data_cleaned.info()
```

```python
data_cleaned.value_counts('ability_2') #gibt die Anzahl der Werte pro Spalte aus, NaN sind nicht dabei
```

```python
data_cleaned.sort_values(['generation','name']) #sortiert nach generation (wenn diese gleich sind, auch noch nach name)
```

```python
data_cleaned.value_counts('type_1').plot.barh()
```
